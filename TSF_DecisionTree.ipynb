# Import the important Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn.datasets as datasets
```

# Import the dataset

```python
df=datasets.load_iris()
```

# Independent Feature

```python
df1=pd.DataFrame(df.data, columns=df.feature_names)
print(df1.head(5))
```

```python
df1.columns
```

```python
df1.describe()
```

# Dependent Variable

```python
y=df.target
print(y)
```

# Plot the features

```python
# Plot for sepal values
plt.rcParams['figure.figsize'] = (12, 8)
fig, ax = plt.subplots()
scatter = ax.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], c = y)
# produce a legend with the unique colors from the scatter
legend = ax.legend(*scatter.legend_elements(),loc="upper right", title="Species")
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('Sepal values', fontsize = 12)
ax.add_artist(legend)
```

```python
# Plot for petal values
plt.rcParams['figure.figsize'] = (12, 8)
fig, ax = plt.subplots()
scatter = ax.scatter(df1['petal length (cm)'], df1['petal width (cm)'], c = y)
# produce a legend with the unique colors from the scatter
legend = ax.legend(*scatter.legend_elements(),loc="upper left", title="Species")
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('Petal values', fontsize = 12)
ax.add_artist(legend)
plt.show()
```

# Apply Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(df1,y)

```

# Test the model for our own data

```python
pred_result = clf.predict([[4.6,3.1,1.5,0.2]])
print("Predicted Result is: ",pred_result)
```

```python
from sklearn import tree
feat_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
cat_names = ['setosa', 'versicolor', 'virginica']
tree.plot_tree(clf, feature_names = feat_names, class_names = cat_names, filled = True);
```
